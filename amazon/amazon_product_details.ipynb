```python
from bs4 import BeautifulSoup
import json
import requests
```

```python
headers={
    "User-Agent":"Mozilla/5.0 (compatible; Googlebot/2.1; +https://www.google.com/bot.html)"
}
url="https://www.amazon.com/HP-Display-i3-1215U-Graphics-15-dy5599nr/dp/B0BVD8LZQL/ref=pd_aw_ci_mcx_pspc_dp_m_m_t_3?pd_rd_w=CSqNA&content-id=amzn1.sym.8601bd6b-5f84-4e74-9236-bbdc6b94ed42&pf_rd_p=8601bd6b-5f84-4e74-9236-bbdc6b94ed42&pf_rd_r=KCYGXQAEB84X82VS0MAF&pd_rd_wg=xL4qz&pd_rd_r=c3b7ef46-f4ee-485b-84c6-f3cf68a667a7&pd_rd_i=B0BVD8LZQL&th=1"
response = requests.get(url,headers=headers)

if response.status_code==200:
    soup=BeautifulSoup(response.text,'html.parser')
    print(soup.prettify())
    
else:
    print("Failed to fetch the webpage")
```

```python
d={}
title= soup.body.find_all('span', id='productTitle')
d['title']=(title[0].text.strip())
price = soup.body.find_all('span', class_='a-price')
d['price']=(price[0].find('span',class_='a-offscreen').text.strip())
div=soup.body.find_all('div',id='prodDetails')
dep=div[0].find_all('tr')

for i in dep:
    d[i.find('th').text.strip()]=i.find('td').text.strip().replace('\n',' ')
print(d)
```

```python

json_data = json.dumps(d)
print(json_data)
```
