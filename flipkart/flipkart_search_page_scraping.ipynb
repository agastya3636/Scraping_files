```python
from bs4 import BeautifulSoup
import requests
import json
from flask import Flask, request
search=input()
search=search.replace(" ","%20")
url=f"https://www.flipkart.com/search?q={search}"
headers={
    "User-Agent":"Mozilla/5.0 (compatible; Googlebot/2.1; +https://www.google.com/bot.html)"
}
```

```python
#fetch the search page
response = requests.get(url,headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    print(soup.prettify())
else:
    print("Failed to fetch the webpage")
```

```python
#extracting the title , price and link of that products
div=soup.body.find_all('div',class_='_1AtVbE col-12-12')
l=[]
for card in div:
    d={}
    title=card.find_all('div',class_='_4rR01T')
    if title:
        d['title']=(title[0].text.strip())
    price=card.find('div',class_='_30jeq3 _1_WHN1')
    if price:
        d['price']=(price.text.strip())
    
    link=card.find('a',class_='_1fQZEK')
    if link:
        #print(link)
        d['link']=('https://www.flipkart.com'+link.get('href'))
    l.append(d)
json_data = json.dumps(l)
print(json_data)
```
